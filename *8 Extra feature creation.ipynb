```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('/Users/didyvanbelkom/Documents/02 MSc DSS/THESIS CODE/Jupiter Notebook/online_retail_II.xlsx', sheet_name='Year 2009-2010')
```

```python
df.head()
```

```python
df["TotalPrice"] = df["Quantity"] * df["Price"]
```

```python
df.head()
```

```python
df.dropna(inplace=True) #erases incomplete observations.

df = df[~df["Invoice"].str.contains("C", na=False)]
df = df[(df['Quantity'] > 0)]
df = df[(df['Price'] > 0)]

df.head()
```

```python
df.describe()
```

```python
df_country = df[['Country', 'Customer ID']]
```

```python
df_country.head()
```

```python
df_country = df_country.drop_duplicates('Customer ID')
```

```python
df_country.describe()
```

```python
df_averagespend = df.groupby(['Customer ID'])['TotalPrice'].mean()
```

```python
df_averagespend.head()
```

```python
df_averagespend.describe()
```

```python
df_averagequant = df.groupby(['Customer ID'])['Quantity'].mean()
```

```python
df_averagequant.head()
```

```python
df1 = pd.read_excel('/Users/didyvanbelkom/Documents/02 MSc DSS/THESIS CODE/Jupiter Notebook/df_complete.xlsx')
```

```python
df1.head()
```

```python
df2 = pd.merge(left=df1, right=df_averagequant, left_on='Customer_ID', right_on='Customer ID')
df2.head()
```

```python
df3 = pd.merge(left=df2, right=df_averagespend, left_on='Customer_ID', right_on='Customer ID')
df3.head()
```

```python
df3.rename(columns = {'TotalPrice_y':'y1 average spend', 'TotalPrice_x':'TotalPrice', 'Quantity':'y1 average quantity'}, inplace = True)
```

```python
df3.head()
df3.describe()
```

```python
df4 = pd.merge(left=df3, right=df_country, left_on='Customer_ID', right_on='Customer ID')
df4.head()
```

```python
df_add_features = df4
```

```python
df_clustering = df_add_features[['recency', 'frequency', 'monetary', 'y1 average quantity', 'y1 average spend', 'Top_10%_spend']]
```
